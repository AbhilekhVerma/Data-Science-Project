# TRAIN TEST SPLIT

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=.25, random_state=56)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# MODEL BUILDING

```python
#LogisticRegression
lr=LogisticRegression()
lr.fit(x_train, y_train)
predlr=lr.predict(x_test)
print(accuracy_score(y_test,predlr))
print(confusion_matrix(y_test,predlr))
print(classification_report(y_test,predlr))
    
    
```

```python
# Logistic Regression Cross Validation

scr=cross_val_score(lr,x,y,cv=5)
print('cross validation score',scr.mean())
```

Accuracy score is 77% and cv score is 75% of Logistic Reression.

```python
# AUC ROC CURVE Logistic Regression
lr.fit(x_train,y_train)
y_pred_proba=lr.predict_proba(x_test)[:,1]
```

```python
y_pred_proba
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred_proba)
fpr
tpr
```

Auc_score of Logistic Regression is 50%.

# HYPER PARAMETER TUNING OF MODEL

```python
from sklearn import datasets
from sklearn.model_selection import GridSearchCV

dt=DecisionTreeClassifier( random_state=0,max_depth=7,min_samples_leaf=13)

grid_param={'criterion':['ginny', 'entropy'],'splitter':['best','random']}
           

gd_sr=GridSearchCV(estimator=dt,
                  param_grid=grid_param,
                  scoring='accuracy',
                  cv=5)

gd_sr.fit(x,y)

best_parameters =gd_sr.best_params_
print(best_parameters)
best_result=gd_sr.best_score_
print(best_result)
```

# FINAL MODEL

```python

final_mod=DecisionTreeClassifier(criterion='entropy',splitter='best')
final_mod.fit(x_train,y_train)
pred=final_mod.predict(x_test)
acc=accuracy_score(y_test,pred)
print(acc*100)
```
